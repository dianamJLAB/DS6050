Our goals are:

1. Write a function to output x/2 if x is even, 3x+1 if x is odd.
1. Create a while loop that terminates when x is 1
1. Measure number of steps and max for any given x
1. Place x, number of steps, and max into a dictionary, then append to a dataframe
1. Repeat for 1 through 100

```python
import numpy as np
import pandas as pd
```

```python
def our_goals(x:int = 0) -> int:
    if x % 2 == 0:
        x = x/2
    else:
        x = 3*x + 1
    return x


inputs = np.arange(1, 100)

outputs = []

for i in inputs:
    outputs.append(our_goals(i))
```

```python
# plot outputs
import matplotlib.pyplot as plt
plt.plot(inputs, outputs)
```
